# Data Analysis
In data analysis, we use different Python packages such as:
- NumPy
- Pandas
- SciPy
- Matplotlib
- Seaborn

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

```python
ages = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26]
```

```python
n = len(ages)
total = sum(ages)
mean = total / n 
print(mean)
```

```python
np_arr = np.array(ages)
```

```python
np_arr.mean()
```

```python
np_arr.min()
```

```python
np_arr.max()
```

```python
np_arr.std()
```

```python
np_arr.var()
```

```python
# y = 2x + 3
```

```python
x = np.arange(10)
y = 2 * x  + 3
print(y)
```

```python
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Linear equation example')
plt.show()
```

```python
x = np.linspace(0, 50, 50)
y = 2 * x  + 3
plt.scatter(x, y)
plt.show()
```

```python
x = np.linspace(0, 50, 50)
y = 2 * x  + 3
plt.plot(x, y)
plt.show()
```

```python
# y = x**2  + 2
x = np.arange(-100, 101)
y = -x **2  + 2 
plt.scatter(x, y)
plt.show()
```

```python
# y = x**2  + 2
x = np.arange(-100, 101)
y = -x **3  + 2 * x ** 2 + 3 * x + 10
plt.scatter(x, y)
plt.show()
```
