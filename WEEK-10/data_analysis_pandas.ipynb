```python
df = pd.read_json('./cats.json')
```

```python
df
```

```python
life_spans = df['life_span']
lst = []
for life_span in life_spans:
    lowest, _, highest = life_span.split()
    average = (int(lowest) + int(highest)) / 2
    lst.append(average)
df['life_span'] = lst
```

```python
# df[['name', 'origin']].groupby('name').sorted()
sorted_data =  df[['name', 'origin']].groupby(['origin']).count()
sorted_data
```

```python
# df = pd.DataFrame({'lab':['A', 'B', 'C'], 'val':[10, 30, 20]})
# ax = df.plot.bar(x='lab', y='val', rot=0)
countries = sorted_data.index
values = [int(value[0]) for value in sorted_data.values]
print(values)
```

```python
fig, ax = plt.subplots()
ax.bar(countries, values)
plt.xticks(rotation=90)
plt.xlabel('Countries of cat origin')
plt.ylabel('Cat breeds')
plt.title('Different cat breeds')
plt.show()
```

```python
# dataframe1.groupby('subjects').size().sort_values(ascending=False)


sorted_df = df[['name', 'origin']].groupby('origin').size().sort_values(ascending=True)
countries = sorted_df.index
print(countries)
values = sorted_df.values
print(values)
```

```python
plt.barh(countries, values)
plt.savefig('cats_barh.png')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
fig.subplots_adjust(bottom=0.5)
ax.bar(countries, values, color='lightgreen')
plt.xticks(rotation=90)
plt.xlabel('Countries of cat origin')
plt.ylabel('Cat breeds')
plt.title('Different cat breeds')
plt.savefig('cats_breeds.png')
plt.show()
```

```python
import matplotlib.pyplot as plt



fig, ax = plt.subplots()
ax.pie(values, labels=countries, autopct='%1.1f%%')
plt.show()
```
